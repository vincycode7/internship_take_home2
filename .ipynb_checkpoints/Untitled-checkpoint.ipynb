### first function to process data

```python
def first_split(trainset=None):
    nwtrain1 = list()
    for i in range(0,len(trainset)):
        try:
            nwtrain1.append(';'.join(trainset.values[i]))
        except:
            try:
                nwtrain1.append(';'.join(trainset.values[i][:4]))
            except:
                try:
                    nwtrain1.append(';'.join(trainset.values[i][:3]))
                except:
                    try:
                        nwtrain1.append(';'.join(trainset.values[i][:2]))
                    except:
                        nwtrain1.append(';'.join(trainset.values[i][:1]))
                        

    return nwtrain1
```

### second function to process data

```python
def sec_split(dat=None):
    dat_nw = []
    for i in range(0,len(dat)):
        chk = []
        for j in dat[i]:
            try:
                chk.append(int(j))
            except:
                try:
                    chk.append(float(j))
                except:
                    try:
                        chk.append(j.split('\"')[1])
                    except:
                        try:
                            chk.append(int(j))
                        except:
                            chk.append(j)

        dat_nw.append(chk)
    return dat_nw
```

### third function to process data

```python
def third_split(data=None):
    for i in data:
        try:
            if len(i) == 22: pass
            elif len(i) == 21:
                if type(i[10]) == str:
                    i.insert(10,np.nan)
                elif i[3] > 8:
                    i.insert(3,np.nan)
                else:
                    i.insert(1,np.nan)
            
            elif len(i) == 20:
                if i[3] > 8:
                
                    if type(i[3]) == str:
                        i.insert(3,np.nan)
                        i.insert(10,np.nan)
                    else:
                        i.insert(1,np.nan)
                        i.insert(3,np.nan)
                elif i[3]<8:
                    i.insert(1,np.nan)
                    i.insert(10,np.nan)
            else:
                i.insert(1,np.nan)
                i.insert(3,np.nan)
                i.insert(10,np.nan)
        except:
            i.insert(1,np.nan)
            i.insert(3,np.nan)
        np.array(i)
    return data
```

### fourth function to process data

```python
def col_swap(trainset=None):
    for i in range(0,len(trainset)):
        if type(trainset[9][i]) != int:
            try:
                trainset[9][i] = int(trainset[9][i])
            except:
                trainset[9][i] = trainset[9][i]
        if type(trainset[4][i]) != int:
            try:
                trainset[4][i] = int(trainset[4][i])
            except:
                try:
                    trainset[4][i] = float(trainset[4][i])
                except:
                    trainset[4][i] = trainset[4][i]
                             
        if type(trainset[4][i]) != int and type(trainset[4][i]) != float:
            hh = trainset[8][i]
            trainset[8][i] = trainset[4][i]
            trainset[4][i] = hh
        
        if type(trainset[9][i]) != int:
            hh = trainset[5][i]
            trainset[5][i] = trainset[9][i]
            trainset[9][i] = hh
    return trainset
```
